```python
import keras_ocr
pipeline = keras_ocr.pipeline.Pipeline()
```

```python
from PIL import Image
```

```python

data = Image.open('data.png')
```

```python
data
```

```python
# keras_ocr.tools.read(['data.png'])
images = [
    keras_ocr.tools.read(img) for img in ['data.png']
]
```

```python
results = pipeline.recognize(images)
```

```python
results
```

```python
# combine the predicted text into a single string
text = ''
for result in results[0]:
    text += result[0] + ' '

text
```
